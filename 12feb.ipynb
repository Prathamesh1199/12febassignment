```python
4: 
    
1."try and else"

Example:

try:
    num = int(input("enter a number:"))
except ValueError:
    print("invalid inout")
else:
    print("yoy entered:",num)


In this example, I have use the try statement to try to convert user input
into an integer using the int() function. If the user enters a non-numeric 
input, a ValueError exception is raised.

The else statement is used to specify the code that should be executed if the
try block is executed successfully, without raising any exceptions. In this
case, I print a message indicating the number entered by the user.

"If the user enters a non-numeric input, the output will be:"
output: Enter a number: abc
        "Invalid input'"

Using the else statement in combination with try and except allows 
to separate error handling code from the main code block, and execute 
different code based on whether an exception was raised or not.



2."finally"

Example:
    
try:
    file = open("example.txt","r")
    lines = file.readlines()
    for line in lines:
        print(line)
except FileNotFoundError:
    print("File Not Found")
finally:
    file.close()


In this example, I have use the try statement to attempt to open a file 
called "example.txt" for reading using the open() function. 
If the file is found, its contents are read into a list of lines using the
"readlines() method".

If the file is not found, a FileNotFoundError exception is raised, and the
except block is executed, printing an error message indicating that the file
was not found.

If the file is found and opened successfully, the contents of the file are
printed to the console. Regardless of whether an exception was raised or not,
the file is closed after the contents have been read.

  Using the "finally" block in combination with 'try and except' allows you to
ensure that certain code is executed regardless of whether an exception was 
raised or not. This can be useful for tasks such as closing files or releasing
resources that should always be cleaned up, even in the event of an error.


3. "Raise"

Example:
    
def divide(x, y):
    if y == 0:
        raise ValueError("Division by zero is not allowed!")
    return x / y

try:
    result = divide(10, 0)
except ValueError as e:
    print(e)
else:
    print(result)


 In this example, I define a function called divide() that takes two
arguments and returns their quotient. I use an if statement to check whether
the second argument y is equal to zero. If it is, I raise a ValueError 
exception with a message indicating that division by zero is not allowed.

then i use a try statement to call the divide() function with arguments
10 and 0. Since the second argument is 0, the ValueError exception is raised.
The except statement catches the ValueError exception and prints the error 
message passed to the raise statement.
The except statement catches the 'ValueError' exception and prints the error 
message passed to the raise statement.

If the second argument to divide() is a non-zero value, the quotient is 
calculated and returned, and the else block is executed, printing the result
to the console.

Using the raise statement allows us to raise our own exceptions in Python,
with a custom message or other information. This can be useful when we need 
to handle specific error conditions that are not covered by the built-in
Python exceptions.


```

```python
5:
    In Python, custom exceptions are user-defined exception classes that we 
can create to raise and handle specific types of errors that are specific to
our code. We need custom exceptions to provide more specific error messages to
the user, which can help them quickly identify the source of the problem and
take appropriate actions to resolve the issue...

Example:
    
class InvalidInputException(Exception):
    pass

def divide(num1, num2):
    if num2 == 0:
        raise InvalidInputException("The second number cannot be zero.")
    else:
        return num1 / num2

try:
    result = divide(10, 0)
except InvalidInputException as e:
    print(e)



In this example, I have define a custom exception called "InvalidInputException"
that inherits from the built-in Exception class.
 then i define a function called divide() that takes two numbers as arguments
and returns their division result. However, if the second number is zero,
i raise the InvalidInputException with a message indicating that the
second number cannot be zero.

then i call the divide() function with two arguments: 10 and 0,
which triggers the InvalidInputException and
prints the error message to the console.

By using custom exceptions, we can provide more specific and informative 
error messages that can help users understand what went wrong and take 
appropriate actions to fix the problem...

```
